# This is an algorithm to convert a 5-point likert based scale to a continuess Guassian fuzzy number. 

```python
#First one must import the necessary data analysis librarires

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random as rnd
#Warnings might become somewhat troublesome so the next import is for convinience.
import warnings
warnings.filterwarnings("ignore")
```

```python
#reading the raw data first.
raw_data=pd.read_csv('Fuzzification.csv')
```

```python
#trimming the unnecessary data away

raw_data.drop(columns='idcode',inplace=True)
```

```python
raw_data.drop(columns=['SMU','SM','PK1','PK2','PK3','Age'],inplace=True)
```

```python
#defining the dictionnaries for TFNS
low_bound={1:0.000,2:0.000,3:0.250,4:0.500,5:0.750}
mid_bound={1:0.000,2:0.250,3:0.500,4:0.750,5:1.000}
up_bound={1:0.250,2:0.500,3:0.750,4:1.000,5:1.000}

#defining the CoG dictionary

CoG={1:((low_bound[1]+4*mid_bound[1]+up_bound[1])/6),
     2:((low_bound[2]+4*mid_bound[2]+up_bound[2])/6),
     3:((low_bound[3]+4*mid_bound[3]+up_bound[3])/6),
     4:((low_bound[4]+4*mid_bound[4]+up_bound[4])/6),
     5:((low_bound[5]+4*mid_bound[5]+up_bound[5])/6)}

```

```python
raw_data_CoG=raw_data.copy()
for col in raw_data_CoG.columns:
    raw_data_CoG[col]=raw_data_CoG[col].map(CoG)
raw_data_CoG
```

```python
import random
Mu={1:0.021,2:0.146,3:0.375,4:0.625,5:0.979}
std={1:0.021,2:0.104,3:0.125,4:0.125,5:0.021}

#defining the random dictionary 
rnd={1:random.normalvariate(Mu[1], std[1]),
    2:random.normalvariate(Mu[2], std[2]),
    3:random.normalvariate(Mu[3], std[3]),
    4:random.normalvariate(Mu[4], std[4]),
    5:random.normalvariate(Mu[5], std[5])}

raw_data_rnd=raw_data.copy()

for col in raw_data_rnd.columns:
    raw_data_rnd[col]=raw_data_rnd[col].apply(lambda x: random.normalvariate(Mu[x], std[x]) if x in Mu else x)
raw_data_rnd
```
